# Random Variable


```python
# Bernoulli 
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from math import factorial , sqrt , pi
p=0.3
n_samples=1000
bern= np.random.binomial(1,p,size=n_samples)
print(bern.mean())
print(bern.var())

mu,sigma= 2.0,1.5
norm_samples = np.random.normal(mu,sigma,size=n_samples)
print("\nNormal sample mean (approax mu):",norm_samples.mean())
print("\nNormal sample (approax sigma^2):",norm_samples.var())

fig,axes = plt.subplots(1,2,figsize=(10,4))
vals,counts = np.unique(bern,return_counts=True)
axes[0].bar(vals,counts/len(bern))
axes[0].set_title('Bernoulli PMF (emperial)')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Probability')

#Normal PDF(emperical histogram = Theoritical PDF)

axes[1].hist(norm_samples,bins=40,density=True)
x = np.linspace(mu-4*sigma,mu+4*sigma,200)
axes[1].plot(x,stats.norm.pdf(x,mu,sigma))
axes[1].set_title("Normal pdf empirical + theoritical ")
plt.tight_layout()
plt.show()
```

# common distance distibution

```python
from scipy import stats
n,p = 10,0.4
k = np.arange(0,n+1)
pmf_binom = stats.binom.pmf(k,n,p) 

print(stats.binom.mean(n,p),stats.binom.var(n,p))

lam = 3.0
k_p = np.arange(0,15)
pmf_pois = stats.poisson.pmf(k_p,lam)
print(stats.poisson.mean(lam),stats.poisson.var(lam))

# fig,ax = plt.subplot(1,2,figsize=(10,4))
# ax[0].stem(k,pmf_binom)
# ax[0].set_title(" binomial ")
plt.stem(k,pmf_binom)
plt.show()

plt.stem(k_p,pmf_pois)
plt.show()
```

```python
#Uniform 
a,b = 0,5
uni_samps = np.random.uniform(a,b,size=10000)
x = np.linspace(a,b,200)

#Expnential
lam = 1.3
exp_samps = np.random.exponential(1/lam,size=10000)
x_exp = np.linspace(0,8,200)

fig,ax = plt.subplots(1,2,figsize=(10,4))
ax[0].hist(uni_samps,bins=10,density=True)
ax[0].plot(x,stats.uniform.pdf(x,a,b-a))
ax[0].set_title("Unform Distribution (emperial+ theoritical)")

ax[1].hist(exp_samps,bins=10,density=True)
ax[1].plot(x_exp,stats.expon.pdf(x_exp,scale=1/lam))
ax[1].set_title(" Exponntial Distribution emperical + theoritical")
plt.show()


xs = np.linspace(-4,4,200)
plt.plot(xs,stats.norm.cdf(xs))
plt.title("Standard normal cdf")
plt.show()
```

# Multiple random variable

```python
mean = [0,0]
cov= [[1.0,0.6],[0.6,1.0]]
samples = np.random.multivariate_normal(mean,cov,size=5000)
x = samples[:,0]
y = samples[:,1]

print("Emperical covariance ",np.cov(x,y)[0,1])
print("Emperical correlation ", np.corrcoef(x,y)[0,1])

plt.figure(figsize=(6,5))
plt.scatter(x[:500],y[:500],s=10,alpha=0.6)
plt.title("Scatter plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# CLT
pop = np.random.exponential(scale=1.0,size=200000)
sample_size=50
n_trials=2000
means =[np.mean(np.random.choice(pop,sample_size)) for _ in range(n_trials)]

plt.figure(figsize=(8,4))
plt.hist(means,bins=10,density=True)

mu_hat = np.mean(means)
sigma_hat = np.std(means)

x = np.linspace(mu_hat-4*sigma_hat,mu_hat+4*sigma_hat,200)
plt.plot(x,stats.norm.pdf(x,mu_hat,sigma_hat))
plt.title("CLI")
plt.show()
```
