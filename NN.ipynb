```python
import numpy as np;

x=np.array([0.5,0.2,0.1])
w=np.array([0.4,-0.7,0.2])
y_sum = np.dot(x,w)
output = y_sum
print("Weighted sum ",y_sum)
print("Output ",output)
```

```python
def biological_neuron(signals,threshold=1.0):
    soma = sum(signals)
    if soma>=threshold:
        return 1
    return 0;
signals =[0.3,0.4,0.5]
print("Neuron Output ",biological_neuron(signals))
    
```

```python
import numpy as np
def artificial_neuron(inputs,weights,bias=0):
    y_sum=np.dot(inputs,weights)+bias
    return y_sum

x=np.array([1,0,1])
y=np.array([0.5,-0.6,0.2])
print(artificial_neuron(x,y,bias=0.1))
```

# Mathematical Functions

```python
import math
def identity(x):
    return x
def step(x):
    return 1 if x>=0 else 0
def threshold(x,theta=0.5):
    return 1 if x>=0.5 else 0
def relu(x):
    return max(0,x)
def sigmoid(x,k=1):
    return 1/(1+math.exp(-k*x))
def bipolar_sigmoid(x):
    return (2/(1+math.exp(-x)))-1
def tanh_n(x):
    return math.tanh(x)

values=[-2,-1,0,0.5,1,2]
print("Input Values ",values)
print("\n--- Identity Function\n")
for v in values:
    print(v,"->",identity(v))
    
for v in values:
    print(v,"->",step(v))   
    
    
```
