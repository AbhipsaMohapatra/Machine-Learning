```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array([[1,2],[2,1],[3,4],[5,7],[6,8],[7,6],[8,5],[4,3]])

plt.scatter(X[:,0],X[:,1])
#plt.plot(a1[:,0],a1[:,1])
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter plot before Clustering")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3,random_state=42)
model = kmeans.fit_predict(X)

plt.scatter(X[:,0],X[:,1],c=model)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='x')

plt.title("K-Means Clustering ")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

# Elbow method

```python
wcss=[]
for k in range(1,X.shape[0]+1):
    km = KMeans(n_clusters=k,random_state=42)
    km.fit(X)
    wcss.append(km.inertia_)
plt.plot(range(1,X.shape[0]+1),wcss,marker='o')
plt.xlabel("no of cluster K")
plt.ylabel("wcss")
plt.title("Elbow Method")
plt.grid(True)
plt.show()
    
    
```
